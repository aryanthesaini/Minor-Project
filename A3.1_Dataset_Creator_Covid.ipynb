```python
import pandas as pd
```

```python
# Creating the data_for positive samples 

FILE_PATH="metadata.csv"
IMAGES_PATH="covidimages"


df=pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

## We can see that there is a coloumn named finding that denotes the disease i.e whether there is covid or not 

## Now we will make a sperate folder only for covid positive xrays


```python
# os is use to iterate over folders and move file into that
import os
import shutil



# Making a folder name covid inside the dataset directory
TARGET_DIR="Dataset/Covid"


# IF this path does not exist then make it i.e make directory(mkdir)
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("covid folder created")
```

## Now our empty covid folder is created so we will pick images of covid dataset and place inside this folder

```python
cnt=0

for(i,row) in df.iterrows():
    if row['finding']=="COVID-19":
        cnt+=1
print(cnt)
```

## i.e we have 521 images of covid positive xray

## The problem here is that some of the xrays have top view while some have side view and so on but we only need the xrays of the front view
## More technically we want that we want posterior -anterior view of xray i.e PA view

```python
cnt=0

for(i,row) in df.iterrows():
    if row['finding']=="COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        image_path=os.path.join(IMAGES_PATH,filename)
        image_copy_path=os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        print("Moving Image ",cnt)
        cnt+=1
print(cnt)
```

## Now our dataset path i.e Dataset/Covid have images of covid x-ray patients
## i.e all the front view or posterior anterior view of xrays

## Now we have to do two things firstly we have to reduce the size of xray images so that our model becomes computational efficient and second we have to make  another folder for normal xrays

```python
## Kaggle dataset is a big dataset with almost 6000 images of normal xray as well 
## but we want to have 50 -50 ratio of the covid and normal images hence we will do image sampling
## half covid images and half normal images


## Randomly choosing 201 (same as the covid xray images) images from the normal dataset

import random
KAGGLE_FILE_PATH="chest_xray/train/NORMAL"
TARGET_NORMAL_DIR="Dataset/Normal"
```

```python
# Getting the names of all the images present in the file path
image_names=os.listdir(KAGGLE_FILE_PATH)
```

```python
len(image_names)
```

```python
# Do the random shuffling of the image names 
random.shuffle(image_names)


for i in range(201):
    image_name=image_names[i]
    image_path=os.path.join(KAGGLE_FILE_PATH,image_name)
    target_path=os.path.join(TARGET_NORMAL_DIR,image_name)
    
    shutil.copy2(image_path,target_path)
    print("copying image ",i)
    
    
    
    
    
    
    
```
